El análisis de datos se presenta como una herramienta fundamental para lograr la toma de decisiones, y en este proyecto nos enfocaremos en la segmentación de clientes mediante técnicas avanzadas de análisis de datos con el uso del lenguaje de programación Python

Importaremos los paquetes esenciales necesarios para esta función y luego leeremos nuestros datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Mall_Customers.csv')
data
```

Visualización del género del cliente


```python
data['Gender'].value_counts()
```

En esto, crearemos un diagrama de barras y un gráfico circular para mostrar la distribución de género en nuestro conjunto de datos de datos de clientes.

```python

plt.hist(data['Gender'], bins = 3)

plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Number of men and women')

plt.show()
```

Del diagrama de barras anterior, observamos que el número de mujeres es mayor que el de hombres. Ahora, visualicemos un gráfico circular para observar la proporción de distribución de hombres y mujeres.

```python
dato = data['Gender'].value_counts()
```

```python
plt.pie(dato, labels=dato.index, autopct='%1.1f%%', startangle=90, colors=['#fb7991', '#3a89c9'])
plt.title('PROMEDIO DE SOBREVIVIENTES POR EL TIPO DE CLASE')
plt.show()
```

Del gráfico anterior, el porcentaje de mujeres es del 56% , mientras que el porcentaje de hombres en el conjunto de datos de clientes es del 44% .

Visualización de la distribución de edades

Primero procederemos tomando un resumen de la variable Edad.

```python
data['Age'].describe()
```

Luego, trazamos un histograma para ver la distribución y trazar la frecuencia de las edades de los clientes. 

```python
sns.histplot(data['Age'], bins=9, color='blue')

plt.xlabel('Age Class')
plt.ylabel('Frequency')
plt.title('Histogram to show the age class count.')

plt.show()
```

La edad promedio de los clientes está entre 30 y 35 años.

```python
# plt.figure(figsize=(7,6))
sns.boxplot(y=data['Age'],color='fuchsia')

# Añadir etiquetas y título
plt.title('Diagrama de Caja')

# Mostrar el diagrama de caja
plt.show()
```

La edad mínima de los clientes es 18 años, mientras que la edad máxima es 70 años.

Análisis de los Ingresos Anuales de los Clientes

```python
plt.figure(figsize=(15,8))

sns.histplot(data['Annual Income (k$)'], bins=13, color='#1b8798')

plt.xlabel('Annual Income Class')
plt.ylabel('Frequency')
plt.title('Histogram for Annual Income')

plt.show()
```

Del análisis descriptivo anterior, concluimos que el ingreso anual mínimo de los clientes es 15 y el ingreso máximo es 137

Las personas que ganan un ingreso promedio de 70 tienen el recuento de frecuencia más alto en nuestra distribución de histograma. El salario medio de todos los clientes es 61,50

```python
sns.kdeplot(data['Annual Income (k$)'], fill=True, color='yellow',alpha = 0.5)
plt.show()
```

En el gráfico de densidad de núcleos que mostramos arriba, observamos que el ingreso anual tiene una distribución normal .

```python
data['Spending Score (1-100)'].describe()
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(x=data['Spending Score (1-100)'],color='red')

# Mostrar el diagrama de caja
plt.show()
```
